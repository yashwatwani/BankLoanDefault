```python
# Correlation matrix - numerical correlation between features
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(30, 25))
mat = df.corr('pearson')
mask = np.triu(np.ones_like(mat, dtype=bool))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(mat, mask=mask, cmap=cmap, vmax=1, center=0, annot = True,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
# Looking at NaN % within the data
# if NaN % in under 20% keep it else discard it
nan = pd.DataFrame(df.isna().sum(), columns = ['Total Missng Values'])
nan['Feature'] = nan.index
nan['Perc(%)'] = (nan['Total Missng Values']/532428)*100
nan = nan[nan['Total Missng Values'] > 0]
nan = nan.sort_values(by = ['Total Missng Values'])
nan['Usability'] = np.where(nan['Perc(%)'] > 20, 'Discard', 'Keep')
nan
```

```python
def bar_chart(df,feature):
    defaulter = df[df['loan_status']==1][feature].value_counts()
    non_defaulter = df[df['loan_status']==0][feature].value_counts()
    df = pd.DataFrame([defaulter,non_defaulter])
    df.index = ['defaulter','non_defaulter']
    df.plot(kind='bar',stacked=False, figsize=(10,5))
```
